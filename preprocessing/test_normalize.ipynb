```python
for x in list(df.columns)[0:1]:
    sns.distplot(list(series[x]))
    plt.show()
    sns.distplot(list(df[x]))
    plt.show()  
    sns.distplot(list(df2[x]))
    plt.show()  
```

```python

stat, p = stats.shapiro(list(df2[list(df2.columns)[0]]))
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
